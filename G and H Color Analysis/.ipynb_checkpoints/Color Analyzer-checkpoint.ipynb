```python
# convert RGB into Hex color
# means we will have one output (hex) vs three diff values (R, G, B)

def rgb_to_hex(rgb_color):
    hex_color = "#"
    for i in rgb_color:
        i = int(i)
        hex_color += ("{:02x}".format(i))
    return hex_color
```
